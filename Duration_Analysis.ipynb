```python
#Transform non-numeric columns into numerical columns
pd.set_option('display.max_columns', 30) 
from sklearn.preprocessing import LabelEncoder
for column in data.columns:
        if data[column].dtype == np.number:
            continue
        data[column] = LabelEncoder().fit_transform(data[column])
data.head(20)
```

```python
T=data.YearsAtCompany
from lifelines import KaplanMeierFitter
# Initiate and fit
kmf = KaplanMeierFitter()
kmf.fit(durations=T, event_observed=data.Attrition)
# Plot the survival function
kmf.survival_function_.plot()
plt.title('Survival Curve estimated with Kaplan-Meier Fitter')
plt.show()
```

```python
# Print survival probabilities at each year
kmf.survival_function_
```

```python
# Plot the survival function with confidence intervals
kmf.plot_survival_function()
plt.show()
```

```python
columns_selected = ['Attrition', 'JobSatisfaction', 'JobInvolvement', 'JobLevel', 'DistanceFromHome', 'PercentSalaryHike', 'TrainingTimesLastYear', 'YearsAtCompany']
data_sel = data[columns_selected]
```

```python
from lifelines import CoxPHFitter
# Initialize and fit the model
coxph = CoxPHFitter()
coxph.fit(data_sel, duration_col='YearsAtCompany', event_col='Attrition')
# Print model summary
coxph.print_summary()
```

```python
# Plot factor effects 
coxph.plot()
```

```python
# Current employees
data_new = data.loc[data['Attrition'] == 0]
# Years at the company of current employees
data_new_obs = data_new['YearsAtCompany']
# Predict survival probabilities 
predicted=(coxph.predict_survival_function(data_new,conditional_after=data_new_obs))*100
predicted
```
